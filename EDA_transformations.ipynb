# <font color='green'>Data Transformations</font>

```python
pt = PowerTransformer(method='yeo-johnson')
y_TSS_transformed = pt.fit_transform(y_TSS.values.reshape(-1, 1)).flatten()
```

```python
y_TSS_trans = pd.DataFrame(y_TSS_transformed, columns=['TSS (mg/L)'])
```

```python
X_TSS_transformers = {}

# Prepare a copy for the transformed data
X_TSS_trans = X_TSS.copy()

# Loop through numeric columns
for col in X_TSS.select_dtypes(include='number').columns:
    pt = PowerTransformer(method='yeo-johnson')
    reshaped = X_TSS[[col]].values  # 2D array
    X_TSS_trans[col] = pt.fit_transform(reshaped)
    X_TSS_transformers[col] = pt  # Save the transformer for later inversion

# To inverse transform a column later:
#col_to_inverse = 'your_column_name'
#original_values = transformers[col_to_inverse].inverse_transform(
    #df_transformed[[col_to_inverse]]
#)
```

```python
for col in X_TSS_trans.columns:
    fig = px.histogram(X_TSS_trans, x=col, title=f'Histogram of {col}')
    fig.show()
```
